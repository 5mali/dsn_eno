```python
import numpy as np
import matplotlib.pyplot as plt
import os
from os.path import dirname, abspath, join
from os import getcwd
import sys
```

```python
AMAX = 10
BMAX = 9250
```

```python
# THIS_DIR = getcwd()
# SDATA_DIR = abspath(join(THIS_DIR, '../' 'solar_data'))  #abspath(join(THIS_DIR, '../../..', 'data'))

lp_action = np.load('./lp/lp_action_TOKYO_2018.npy').flatten()/AMAX
lp_batt = np.load('./lp/lp_battery_TOKYO_2018.npy').flatten()/BMAX

nn_action = np.load('nn_action_TOKYO_2018.npy')
nn_action = (nn_action+1)/AMAX
nn_batt = np.load('nn_battery_TOKYO_2018.npy')

henergy = np.load('henergy_TOKYO_2018.npy')
```

```python
day_start = 270*24
day_end = 290*24

fig = plt.figure(figsize=(24,6))
fig.suptitle('TOKYO 2018', fontsize=15)

ax = fig.add_subplot(211)
ax.set_title('\n\nLP OPTIMIZATION')
ax.plot(lp_batt[day_start:day_end],'r')
ax.plot(lp_action[day_start:day_end],'b',alpha=0.5)
ax.plot(henergy[day_start:day_end],'k',alpha=0.7)
ax.set_ylim([0,1.0])
ax.legend(('Battery','Duty Cycle','Solar Energy'))

ax1 = fig.add_subplot(212)
ax1.set_title('\n\nNN OPTIMIZATION')
ax1.plot(nn_batt[day_start:day_end],'r')
ax1.plot(nn_action[day_start:day_end],'b',alpha=0.5)
ax1.plot(henergy[day_start:day_end],'k',alpha=0.7)
ax1.set_ylim([0,1.0])
ax1.legend(('Battery','Duty Cycle','Solar Energy'))

fig.tight_layout()

plt.show()
```
